```python
file_path = 'RCRemap_00000001682065033360_W1920_Speed_0.jpg'
```

```python
import os
import matplotlib.pyplot as plt
import matplotlib.patches as pat
import cv2 as cv
from PIL import Image, ImageDraw
```

### 方法一 使用matplotlib

```python
img = plt.imread(file_path)
fig, ax = plt.subplots(1)
ax.imshow(img)
# Rectangle 坐标的参数格式为左上角（x, y）, width, height
rec = pat.Rectangle((522, 255), 611, 34, linewidth=1, edgecolor='r', facecolor='None')
ax.add_patch(rec)
plt.imshow(img)
plt.xticks([]), plt.yticks([])
plt.show()
fig.savefig('im1_rec.jpg')
```

### 方法二 使用PIL

```python
img = Image.open(file_path)
a = ImageDraw.ImageDraw(img)
# rectangle 坐标的参数格式为左上角（x1, y1），右下角（x2, y2）
a.rectangle(((522, 255), (522 + 611, 255 + 34)), fill=None, outline='red', width=1)
img.save('img2_rec.jpg')
img.show()
```

### 方法三 使用opencv

```python
img = cv.imread(file_path)
# rectangle 坐标的参数格式为左上角（x1, y1），右下角(x2, y2)
cv.rectangle(img, (522, 255), (522 + 611, 255 + 34), (0, 0, 255), 1)
cv.namedWindow('img', cv.WINDOW_NORMAL)
cv.imshow('img', img)
cv.waitKey(0)
cv.imwrite('img3_rec.jpg', img)
cv.destroyAllWindows()
```
